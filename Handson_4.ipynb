## 6.1 Prime Number Checker
Create a program that checks whether a number is a prime number and displays its factors if it is not a prime number.

### Console:
```powershell
Prime Number Checker

Please enter an integer between 1 and 5000: 5
5 is a prime number.

Try again? (y/n): y

Please enter an integer between 1 and 5000: 6
6 is NOT a prime number.
It has 4 factors: 1 2 3 6

Try again? (y/n): y

Please enter an integer between 1 and 5000: 200
200 is NOT a prime number.
It has 12 factors: 1 2 4 5 8 10 20 25 40 50 100 200

Try again? (y/n): n

Bye!
```

### Specifications:
- A prime number is divisible by two factors (1 and itself). For example, 7 is a prime number because it is only divisible by 1 and 7.
- If the user enters an integer that's not between 1 and 5000, the program should display an error message.
- If the number is a prime number, the program should display a message.
If the number is not a prime number, the program should display a message. Then, it should display the number of factors for the number and a list of those factors.
- Store the factors for each number in a list.
- Use functions to organize the code for this program.


````python
### CODE HERE ###
def get_factors(num):
    factors = []        #List

    for i in range(1, num + 1):
        if num % i == 0:
            factors.append(i)

    return factors
#////////////////////////////////////////////////////////

def is_valid(num):
    if num >= 1 and num <= 5000:               #checks number range 1 - 5000
        return True
    else:
        return False
#////////////////////////////////////////////////////////

def main():
    print("Prime Number Checker")
    print()     #space

    choice = "y"
    #choice = "x"
    while choice == "y":
        num = int(input("Please enter an integer between 1 and 5000: "))

        if is_valid(num) == False:
            print("Number must be between 1 and 5000.")
        else:
            factors = get_factors(num)

            if len(factors) == 2:
                print(num, "is a prime number.")
            else:
                print(num, "is NOT a prime number.")
                print("It has", len(factors), "factors:", end=" ")

                for f in factors:
                    print(f, end=" ")

                print()

        print()
        choice = input("Try again? (y/n): ")

    print("Bye!")


main()
````

## 6.2 - Contact Manager
Create a program that a user can use to manage the primary email address and phone number for a contact.

### Console

### Specifications:
- Use a list of lists to store the data for the contacts. Provide starting data for two or more contacts.
- For the `view` and `del` commands, display an error message if the user enters and invalid contact number.
- When you exit the program, all changes that you made to the contact list are lost.




````python
### CODE HERE ###
def display_menu():
    print("COMMAND MENU")
    print("list - Display all contacts")
    print("view - View a contact")
    print("add - Add a contact")
    print("del - Delete a contact")
    print("exit - Exit program")
    print()
#////////////////////////////////////////////////////////

def list_contacts(contact_list):
    if len(contact_list) == 0:
        print("There are no contacts in the list.")
    else:
        number = 1
        for contact in contact_list:
            print(number, contact[0])
            number = number + 1
#////////////////////////////////////////////////////////
                                                            #prints contacts the order they are on contact_list
def view_contact(contact_list):
    number = int(input("Number: "))

    if number < 1 or number > len(contact_list):
        print("Invalid contact number.")
    else:
        contact = contact_list[number - 1]
        print("Name:", contact[0])
        print("Email:", contact[1])
        print("Phone:", contact[2])
#////////////////////////////////////////////////////////
                                                            #adds a contact to the contact_list
def add_contact(contact_list):
    name = input("Name: ")
    email = input("Email: ")
    phone = input("Phone: ")

    contact = [name, email, phone]
    contact_list.append(contact)                 #append makes changes to list

    print(contact[0], "was added.")
#////////////////////////////////////////////////////////
                                                            #deletes a contact from the contact_list
def delete_contact(contact_list):
    number = int(input("Number: "))

    if number < 1 or number > len(contact_list):
        print("Invalid contact number.")
    else:
        contact = contact_list.pop(number - 1)             #.pop removes list item
        print(contact[0], "was deleted.")
#////////////////////////////////////////////////////////

def main():
    contact_list = [["Guido van Rossum", "guido@python.org", "123-456-7890"],              #the main list for contacts
                    ["Eric Idle", "eric@ericidle.com", "987-654-3210"]]

    print("Contact Manager")
    print()     #space
    display_menu()

    command = input("Command: ")                #enter one of the commands for a specific function

    while command != "exit":                    #reminder "!=" is not equal to
        if command == "list":
            list_contacts(contact_list)
        elif command == "view":
            view_contact(contact_list)
        elif command == "add":
            add_contact(contact_list)
        elif command == "del":
            delete_contact(contact_list)
        else:
            print("Not a valid command. Please try again.")

        print()
        command = input("Command: ")

    print("Bye!")


main()
````
